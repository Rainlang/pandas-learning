```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(columns = ['age', 'score'])
df['score'] = np.random.randint(100, size=(10))
df['age'] = np.random.randint(25, size=(10))
print(df)
```

```python
df.plot(x='age', y='score', kind='scatter')
```

```python
# 对df按年龄进行排序
sorted_df = df.sort_values(by = ['age'])
sorted_df
```

```python
sorted_df.plot(x = 'age', y = 'score')
```

```python
# 将sorted_df存入csv文件
sorted_df.to_csv('pd_csv', index=False)
# load a dataframe from csv file
load_df = pd.read_csv('pd_csv')
load_df
```
