```python
# dataframe is more like a 2-dimension array
import pandas as pd
```

```python
data = [['Google', 10], ['Runoob', 12], ['Wiki', 20]]
df = pd.DataFrame(data, columns =['Site', 'Age'])
df
```

```python
# 也能用字典来创建
data = {'Site':['Google', 'Runoob', 'Wiki'], 'Age':[10, 12, 20]}
df = pd.DataFrame(data)
df
```

```python
# 用ndarrays也能创建
import numpy as np
ndarrays_data = np.array([
    ['Google', 10],
    ['Runoob', 12],
    ['Wiki', 20]
])
df = pd.DataFrame(ndarrays_data, columns = ['Site', 'Age'])
df
```

```python
# 用loc属性返回特定行的数据
df.loc[0]
```

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

# 数据载入到 DataFrame 对象
df = pd.DataFrame(data, index=range(1, 4)) # 设置索引，从1开始计数
print(df)
print(df.loc[1]) #返回数据类型其实就是一个series
```

```python
# 可以返回多行数据
print(df.loc[[1, 2]]) #返回值数据其实就是一个dataframe
```

```python
# 通过列表创建
df = pd.DataFrame([[1, 2, 3], [4, 5, 6], [7, 8, 9]],
                  columns=['Column1', 'Column2', 'Column3'])
df
```

```python
# 通过series创建
s1 = pd.Series(['Alice', 'Bob', 'Jack'])
s2 = pd.Series([12, 23, 34])
s3 = pd.Series(['female', 'male', 'male'])
df = pd.DataFrame({'Name':s1, 'Age':s2, 'Gender':s3}) # 将series组成一个字典再进行构造
df
```

```python
# DataFrame 的属性和方法
print(df.shape)     # 形状
print(df.columns)   # 列名
print(df.index)     # 索引
print(df.head(2),"\n")    # 前2行数据，默认是前 5 行
print(df.tail(2),"\n")    # 后2行数据，默认是后 5 行
print(df.info(),"\n")    # 数据信息

print(df['Age'].describe(),"\n")# 描述统计信息
print(df['Age'].mean(),"\n")    # 求平均值
print(df['Age'].sum(),"\n")     # 求和
```

```python
#访问元素
print(df.Name)
print(df['Name'])
print(df.loc[:1,'Name']) # 根据名字取列
print(df.iloc[:,1]) # 根据索引取列
print(df.loc[0, 'Name'])
```

```python
# 修改DataFrame
df = pd.DataFrame(
    [[2, 43, 43], [324, 34, 23], [30, 435, 52]],
    columns = ['c1', 'c2', 'c3']
)
df
```

```python
# 添加新列
df['c4'] = [3, 6, 9]
df
```

```python
# 修改列数据
df['c1'] = [7, 8, 9]
df
```

```python
# 修改行
df.loc[0] = [1, 2, 3, 4]
df
```

```python
# 添加新行
df.loc[3] = [8, 89, 0, 81]
df
```

```python
# concat 方法添加新行
new_row = pd.DataFrame([[4, 7]], columns=['c1', 'c2']) # 先构造一个只有一行的df 
df = pd.concat([df, new_row], ignore_index=True)
df
```

```python
# 删除
df_dropped = df.drop('c1', axis=1)
df_dropped
```

```python
df_dropped = df.drop(0)
df_dropped
```

```python
# 横向合并
new_df = pd.merge(df, df_dropped, on='c4')
new_df
```
